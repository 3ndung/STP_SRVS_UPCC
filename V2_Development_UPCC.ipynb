```
data = data.fillna(method='ffill')
data['SOC Code'] = data['SOC Code'].astype(int)

data[['POC','City','Name']]

for x in range(len(data)):
    data['POC'][x] = data['POC'][x].split('[')[0]
    data['City'][x] = data['City'][x].split('[')[0]


def getel(ml, pos):
    return ml[pos]

data['POC'] = data['POC'].str.split('[').apply(getel,pos=0)
data['City'] = data['City'].str.split('[').apply(getel,pos=0) 

data['POC'] = data['POC'].str.split('\n').apply(getel,pos=0)
data['City'] = data['City'].str.split('\n').apply(getel,pos=0)  

data[['POC','City','Name','SID','SOC Code']]

data = data.fillna(method='ffill')
data['SOC Code'] = data['SOC Code'].astype(int)




data[['POC','City','Name']]

data['Merge_flag']=0

perm = permutations(range(len(data)), 2)  

for i,j in perm:
    if data['POC'][int(i)] == data['POC'][int(j)] and data['City'][int(i)] == data['City'][int(j)]:
        data['Merge_flag'][i] = 1
        
        
if 1 in data['Merge_flag']:
    data = data.groupby(['POC','City']).agg({'SOC Code':'first','POC':'first','City':'first','Name':'_'.join})
    ID = [x for x in range(len(data))]
    data['ID'] = ID
    data.set_index('ID',inplace=True)        
    
data[['POC','City','Name']]    
```
